```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans  # Import KMeans

sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)  # Now KMeans is defined
    kmeans.fit(df_pca)
    sse.append(kmeans.inertia_)

# ... (rest of your code)
```

## Evaluation
Evaluate the clustering result using appropriate metrics.


```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score # Import the silhouette_score function

# ... (Your existing code to fit the KMeans model)

# Get the cluster labels
labels = kmeans.labels_

silhouette_score = silhouette_score(df_pca, labels)
print(f"Silhouette Score: {silhouette_score}")
```
